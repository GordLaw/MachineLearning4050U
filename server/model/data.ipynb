```python
import pandas as pd
import numpy as np
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("nikatomashvili/steam-games-dataset")

print("Path to dataset files:", path)
```

```python
file = path + "/merged_data.csv"
data = pd.read_csv(file)
data = data.dropna()
data
```

```python
print(data["All Reviews Summary"].unique())
print(data["Recent Reviews Summary"].unique())
```

```python
features = data[["Title", "Developer", "Original Price", "Recent Reviews Summary", "All Reviews Summary", "Popular Tags", "Game Features"]]
features["Popular Tags"] = features.loc[:, "Popular Tags"].str.replace("'","", regex=True)
features["Popular Tags"] = features.loc[:, "Popular Tags"].str.replace(r'\[|\]',"", regex=True)
features["Game Features"] = features.loc[:, "Game Features"].str.replace("'","", regex=True)
features["Game Features"] = features.loc[:, "Game Features"].str.replace(r'\[|\]',"", regex=True)
features
```
